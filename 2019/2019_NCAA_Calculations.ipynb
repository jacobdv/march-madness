```python
def remove_NCAA(school):
    school = school.replace(" NCAA","")
    return school
```

```python
def comparison(t_one,t_two):
    result = pd.DataFrame()
    result["School"] = [t_one,t_two]
    result["Points"]
    return result
```

```python
# Pulls Averages for Comparison
avg_w_pct = round(100 * (st.mean(ncaa["W/L%"])),1)
# avg_pace = round(st.mean(ncaa_adv["Pace"]),2)
avg_sos = round(st.mean(ncaa["SOS"]),2)
```

```python
# Key Stats
cs = pd.DataFrame()

# Pulls Additional Stats
cs["School"] = ncaa["School"]
cs["ORS%"] = round(100 * (ncaa["ORB"] / (ncaa["ORB"] + (ncaa_opp["TRB"] - ncaa_opp["ORB"]))),2)
cs["DRS%"] = round(100 * (ncaa["TRB"] - ncaa["ORB"]) / ((ncaa["TRB"] - ncaa["ORB"]) + ncaa_opp["ORB"]),2)
cs["TRS%"] = cs["ORS%"] + cs["DRS%"]
cs["A%"] = round(100* (ncaa["AST"] / ncaa["FG"]),2)

# Filter Stats
cs["W%"] = round(100 * (ncaa["Wins"]/ncaa["Games"]),1)
cs["SOS"] = ncaa["SOS"]

# Filter - Removes schools with losing seasons and below average schedule strengths.
#cs = cs[cs["W%"] >= avg_w_pct]
#cs = cs[cs["SOS"] >= avg_sos]

# Returns to non-filter stats.
cs = cs[['School','ORS%','DRS%','TRS%','A%']]

# Sorts to Display
school_list = []
for school in cs["School"]:
    school = remove_NCAA(school)
    school_list.append(school)
cs["School"] = school_list
#cs.sort_values("TRS%", ascending=False)
cs.sort_values("School")
```

```python
# Algorithms
algorithms = pd.DataFrame()
# Base Data
algorithms["School"] = ncaa["School"]

# Removes 'NCAA' from school names.
algorithms["School"] = algorithms["School"].str.replace('NCAA','') 
#algorihms["Crit"] = 

# Version One: (PPG * (EFG%) / Opp. PPG * (Opp. EFG%) / (200 - (TRB%)))
algorithms["Version One"] = round(1000 * ((ncaa["PPG"] * ncaa_adv["EFG%"]) / (ncaa_opp["PPG"] * ncaa_opp_adv["EFG%"])) / (200 - ncaa_adv["TRB%"]), 2)
# Version Three:
algorithms["Version Three*"] = round((ncaa["Team Points"]/ncaa["FGA"])*ncaa_adv["Pace"],2)
# Sorts DataFrame by uncommented column.
algorithms.sort_values("School", ascending = True)
```

```python
# Image Creation -  All Teams
cs.sort_values("School", ascending=True)
x = cs["RS%"] # Change this one.
y = ncaa["Wins"]
plt.scatter(x,y)
slope,intercept,rvalue,pvalue,stderr = stats.linregress(x,y)
regression_line = slope * x + intercept
equation = f"y = {round(slope,2)}x + {round(intercept,2)}"
plt.plot(x, regression_line)
plt.ylabel("Wins")
plt.xlabel("RS%")
plt.title(f"{equation}   |   R: {round(rvalue,2)}")
plt.show()
```
